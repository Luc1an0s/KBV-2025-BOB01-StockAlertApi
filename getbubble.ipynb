```python
import requests
import pandas as pd

# URL do endpoint
url = "https://appbobinaskbv.bubbleapps.io/version-live/api/1.1/wf/cortes"

# Parâmetros fixos
params = {
    "loja": 33,
    "dataini": "2024-01-01T00:00:00.000Z",
    "datafim": "2025-09-15T00:00:00.000Z",
    "pagina": 1
}

# Lista para armazenar todos os cortes
todos_os_cortes = []

# Primeiro request para descobrir quantas páginas existem
response = requests.get(url, params=params)
data = response.json()

# Total de páginas
total_paginas = data['response']['paginas']

# Adiciona os cortes da primeira página
todos_os_cortes.extend(data['response']['cortes'])

# Loop pelas demais páginas
for pagina in range(2, total_paginas + 1):
    params['pagina'] = pagina
    response = requests.get(url, params=params)
    if response.status_code == 200:
        dados_pagina = response.json()
        cortes_pagina = dados_pagina['response']['cortes']
        todos_os_cortes.extend(cortes_pagina)
    else:
        print(f"Erro na página {pagina}: {response.text}")
        break

# Convertendo tudo para DataFrame
df = pd.DataFrame(todos_os_cortes)

# Exibindo os primeiros registros
print(df.head())
```
